````python
# read gemini_outputs.txt file and parse the python code blocks from it and write it to a new file in the code directory
import os
with open("gemini_outputs.txt", "r") as f:
    lines = f.readlines()
code_blocks = []
in_code_block = False
current_code_block = []
for line in lines:
    if line.startswith("```python"):
        in_code_block = True
    elif line.startswith("```") and in_code_block:
        in_code_block = False
        code_blocks.append("".join(current_code_block))
        current_code_block = []
    elif in_code_block:
        current_code_block.append(line)

# write the code blocks to new files
for i, code_block in enumerate(code_blocks):
    os.makedirs(f"code/problem-{i}", exist_ok=True)
    with open(f"code/problem-{i}/gemini_scot.py", "w") as f:
        f.write(code_block)
````
